```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('bikes_rent.csv')
df.head()
#     
def plot_relationships(df, attributes, target):
    fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 15))
    axes = axes.ravel()
    
    for idx, attr in enumerate(attributes):
        df.plot(kind='scatter', x=attr, y=target, alpha=0.5, ax=axes[idx], title=f'{attr} vs {target}')
    
    plt.tight_layout()
    plt.show()

#     
attributes = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 
              'temp', 'atemp', 'hum', 'windspeed(mph)', 'windspeed(ms)']

#  
target = 'cnt'

#  
plot_relationships(df, attributes, target)
```
